```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "C:/Users/mackd/OneDrive/Desktop/Data_Visualization/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "C:/Users/mackd/OneDrive/Desktop/Data_Visualization/Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete= pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID"])

# Display the data table for preview
study_data_complete.head()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse

last_timepoint = study_mice_cleaned_df.groupby(["Mouse ID"])["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
cleaned_tp_df = pd.merge(last_timepoint,study_mice_cleaned_df,on=("Timepoint","Mouse ID")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
list_tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
 # Locate the rows which contain mice on each drug and get the tumor volumes
for drug in drug_list:
    tumor_volume = cleaned_tp_df.loc[cleaned_tp_df["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    list_tumor_vol.append(tumor_volume)
    
    
    # add subset
    quart = tumor_volume.quantile([.25, .5, .75])
    iqr = quart[.75] - quart[.25]
    lower_bound = quart[.25] - (1.5 * iqr)
    upper_bound = quart[.75] + (1.5 * iqr)
    
    # Determine outliers using upper and lower bounds

    tumor_outliers= tumor_volume[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]
    print(f"{drug}'s potential outliers: {tumor_outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
flierprops=dict(markerfacecolor="blue", markersize=15)

plt.boxplot(list_tumor_vol,labels=drug_list, flierprops=flierprops)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

data = pd.DataFrame({'Weight': weight_average, 'Volume': vol_average})

# Calculate correlation and regression
correlation = st.pearsonr(data['Weight'], data['Volume'])
slope, intercept, rvalue, pvalue, stderr = st.linregress(data['Weight'], data['Volume'])
data['Regression'] = data['Weight'] * slope + intercept

# Output correlation
print(f"The correlation between mouse weight and the average tumor volume is: {correlation[0]:.2f}")

# Plotting
plt.scatter(data['Weight'], data['Volume'], label='Data Points')
plt.plot(data['Weight'], data['Regression'], color='red', label='Regression Line')
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm³)")
plt.legend()
plt.show()
```
