```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "C:/Users/mackd/OneDrive/Desktop/Data_Visualization/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "C:/Users/mackd/OneDrive/Desktop/Data_Visualization/Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete= pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID"])

# Display the data table for preview
study_data_complete.head()
```

```python
# Checking the number of mice.
mice_count=mouse_metadata["Mouse ID"].count()

mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = study_results[study_results.duplicated(subset=['Mouse ID', 'Timepoint'])]
duplicate_mice_ids = duplicate_mice['Mouse ID'].unique()

duplicate_mice_ids
```

```python
# Optional: Get all the data for the duplicate mouse ID.
duplicate_mouse=study_data_complete.loc[study_data_complete["Mouse ID"]=="g989"]
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
study_mice_cleaned_df = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_mice_ids) == False]
study_mice_cleaned_df                                    
```

```python
# Checking the number of mice in the clean DataFrame.
len(study_mice_cleaned_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame

mean=study_mice_cleaned_df.groupby("Drug Regimen").mean(numeric_only=True)["Tumor Volume (mm3)"]
median= study_mice_cleaned_df.groupby("Drug Regimen").median(numeric_only=True)["Tumor Volume (mm3)"]
var= study_mice_cleaned_df.groupby("Drug Regimen").var(numeric_only=True)["Tumor Volume (mm3)"]
stan_dev=study_mice_cleaned_df.groupby("Drug Regimen").std(numeric_only=True)["Tumor Volume (mm3)"]
sem=study_mice_cleaned_df.groupby("Drug Regimen").sem(numeric_only=True)["Tumor Volume (mm3)"]

regimen_summary_df=pd.DataFrame({"Mean Tumor Value": mean,
                                 "Median Tumor Value": median,
                                 "Variance": var,
                                 "Tumor Volume Standard Deviation": stan_dev,
                                 "Tumor Value SEM": sem})
regimen_summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

#stats
mean = study_mice_cleaned_df.groupby("Drug Regimen").mean(numeric_only=True)["Tumor Volume (mm3)"]
med = study_mice_cleaned_df.groupby("Drug Regimen").median(numeric_only=True)["Tumor Volume (mm3)"]
var = study_mice_cleaned_df.groupby("Drug Regimen").var(numeric_only=True)["Tumor Volume (mm3)"]
stan_dev = study_mice_cleaned_df.groupby("Drug Regimen").std(numeric_only=True)["Tumor Volume (mm3)"]
sem = study_mice_cleaned_df.groupby("Drug Regimen").sem(numeric_only=True)["Tumor Volume (mm3)"]

#dataframe
regimen_summary_df=pd.DataFrame({"Tumor Value Mean": mean,
                                 "Tumor Value Median": med,
                                 "Tumor Value Variance": var,
                                 "Tumor Value Standard Dev": stan_dev,
                                 "Tumor Value SEM": sem})

#test
regimen_summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

study_mice_cleaned_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean", "median", "var", "std", "sem"]})
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
drug_bar = study_mice_cleaned_df.groupby("Drug Regimen")
drug_timepoint = drug_bar["Timepoint"].count().sort_values(ascending=False)

# Create a bar plot
mouse_timeplot_bar = drug_timepoint.plot(kind="bar")

# Set labels for the plot
mouse_timeplot_bar.set_xlabel("Drug Regimen")
mouse_timeplot_bar.set_ylabel("# of Observed Mouse Timepoints")

#Plot formatting
plt.tight_layout()
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.

#Defining x and y axises
x = study_mice_cleaned_df["Drug Regimen"].value_counts()

# Plot the bar chart
plt.bar(x.index, x)
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")
plt.xticks(rotation="vertical")
plt.tight_layout()
plt.show()

```

```python
# Generate a pie chart, using Pandas, showing the distribution of unique female versus male mice used in the study

# Get the unique mice with their gender
study_mice_cleaned_df["Sex"].value_counts().plot.pie(y="Mouse ID",ylabel="Count",autopct="%1.1f%%",)

# Make the pie chart
plt.show()
```

```python
# Generate a pie chart, using pyplot, showing the distribution of unique female versus male mice used in the study

# Get the unique mice with their gender
data=study_mice_cleaned_df["Sex"].value_counts()

# Make the pie chart
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.ylabel("Count")
plt.show()


```
